```python
import numpy as np
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import normalize
```

```python
df= pd.read_csv("Creditcard_data.csv")
df.head()
df=df.iloc[:,1:]
```

```python
!pip uninstall scikit-learn
!pip install scikit-learn
```

```python
Amount=normalize([df['Amount']])[0]
df['Amount']=Amount
```

```python
df.head()
```

```python
df=pd.DataFrame(df,columns=df.columns)
```

```python
df['Class'].value_counts()
```

```python
y=df.iloc[:,-1]
x=df.iloc[:,:-1]
x
y
```

```python
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(x,y)
y_resampled.value_counts()
```

```python
new_df=pd.concat([X_resampled,y_resampled],axis=1)
```

```python
new_df.head()
new_df['Class'].value_counts()
```

```python
## Simple random sampling
##
sampled_df_1 = new_df.sample(frac=0.7, random_state=42)
sampled_df_1
```

```python
#systematic sampling
interval = 5

# perform systematic sampling
sampled_df_2 = new_df.iloc[::interval]
sampled_df_2
```

```python
##stratified sampling

strata = new_df.groupby('Class')

# sample 2 rows from each stratum
sampled_df_3 = strata.apply(lambda x: x.sample(frac=0.6))

sampled_df_3
```

```python
#cluster sampling

import numpy as np
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=12, random_state=42).fit(new_df)
cluster_assignments = kmeans.labels_

selected_clusters = [0, 1, 4, 7, 8]
cluster_series = pd.Series(cluster_assignments)
sampled_df_4 = new_df[cluster_series.isin(selected_clusters)]


sampled_df_4
```

```python
#convenience sampling
sample_head=new_df.head(300)
sample_tail=new_df.tail(300)
sampled_df_5=pd.concat([sample_head,sample_tail])
sampled_df_5
```
